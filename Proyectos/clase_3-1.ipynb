# Variables

### Ejemplos utilizando Variables

```python
# Ejemplo 1: Definición de variables
nombre = "Juan"
edad = 28
altura = 1.75
es_estudiante = True

print("Nombre:", nombre)
print("Edad:", edad)
print("Altura:", altura)
print("¿Es estudiante?", es_estudiante)

# Ejemplo 2: Asignación múltiple
a, b, c = 5, 10, 15
print("a:", a, "b:", b, "c:", c)

# Ejemplo 3: Actualización de variables
contador = 0
contador = contador + 1
print("Contador actualizado:", contador)

# Ejemplo 4: Variables de diferentes tipos
x = 3           # int
y = 2.5         # float
z = "Python"    # str
w = False       # bool

print(type(x), type(y), type(z), type(w))

# Ejemplo 5: Uso de variables en operaciones
resultado = a + b * c
print("Resultado de a + b * c:", resultado)
```

### Nomenclatura para las variables

```python
# Ejemplo de nomenclatura 'snake_case'
numero_de_estudiantes = 30
promedio_altura = 1.68

# Ejemplo de nomenclatura 'camelCase'
numeroDeProfesores = 5
promedioEdad = 35

# Ejemplo de nomenclatura 'PascalCase'
TotalAlumnos = 100
PromedioCalificacion = 8.7

# Ejemplo de nomenclatura 'UPPER_CASE' (usualmente para constantes)
PI = 3.1416
MAXIMO_INTENTOS = 3

print("Snake case:", numero_de_estudiantes, promedio_altura)
print("Camel case:", numeroDeProfesores, promedioEdad)
print("Pascal case:", TotalAlumnos, PromedioCalificacion)
print("Upper case:", PI, MAXIMO_INTENTOS)
```

## Tipos de Datos Basicos

```python
# Ejemplo 1: Uso de los tipos básicos
numero_entero = 42          # int
numero_decimal = 3.14       # float
texto = "Hola mundo"        # str
es_valido = False           # bool
valor_nulo = None           # None

print(type(numero_entero), numero_entero)
print(type(numero_decimal), numero_decimal)
print(type(texto), texto)
print(type(es_valido), es_valido)
print(type(valor_nulo), valor_nulo)

# Ejemplo 2: Operaciones y comprobaciones
suma = numero_entero + int(numero_decimal)
mensaje = texto + "!"
es_mayor = numero_entero > 10
sin_valor = valor_nulo is None

print("Suma:", suma)
print("Mensaje:", mensaje)
print("¿Es mayor que 10?:", es_mayor)
print("¿Sin valor?:", sin_valor)
```

### Conversion de Tipos

```python
# Ejemplo 1: Conversión de tipos (casting explícito)
entero = 7
decimal = float(entero)  # int a float
texto_numero = str(entero)  # int a str

print("int a float:", decimal)
print("int a str:", texto_numero)

# Ejemplo 2: Casting explícito de float a int (pierde decimales)
numero_decimal = 3.14
entero_desde_float = int(numero_decimal)
print("float a int:", entero_desde_float)

# Ejemplo 3: Redondeo de números decimales
numero_decimal = 3.14159
redondeado = round(numero_decimal, 2)  # Redondea a 2 decimales
print("Redondeado a 2 decimales:", redondeado)

# Ejemplo 4: Conversión de str a int y float
texto_entero = "123"
texto_decimal = "45.67"
entero_desde_str = int(texto_entero)
decimal_desde_str = float(texto_decimal)
print("str a int:", entero_desde_str)
print("str a float:", decimal_desde_str)
```

## Operadores esenciales

```python
# Operadores aritméticos
suma_aritmetica = a + b
resta_aritmetica = c - a
multiplicacion = b * c
division = c / a
modulo = c % b
potencia = a ** 2

print("Suma:", suma_aritmetica)
print("Resta:", resta_aritmetica)
print("Multiplicación:", multiplicacion)
print("División:", division)
print("Módulo:", modulo)
print("Potencia:", potencia)

# Operadores de comparación
mayor_que = edad > promedioEdad
menor_que = altura < promedio_altura
igual_que = PI == 3.1416
diferente_que = nombre != "Pedro"

print("Mayor que:", mayor_que)
print("Menor que:", menor_que)
print("Igual que:", igual_que)
print("Diferente que:", diferente_que)

# Operadores lógicos
and_logico = es_estudiante and es_valido
or_logico = es_estudiante or es_valido
not_logico = not es_estudiante

print("AND lógico:", and_logico)
print("OR lógico:", or_logico)
print("NOT lógico:", not_logico)

# Operadores de asignación compuesta
contador += 2
a *= 3
b -= 4
c //= 5

print("Contador (+= 2):", contador)
print("a (*= 3):", a)
print("b (-= 4):", b)
print("c (//= 5):", c)

# Operadores de pertenencia
pertenece = "Juan" in nombre
no_pertenece = "Pedro" not in nombre
letra_en_texto = "H" in texto

print("'Juan' in nombre:", pertenece)
print("'Pedro' not in nombre:", no_pertenece)
print("'H' in texto:", letra_en_texto)
```

# Texto y Formato

```python
# Ejemplo de slicing de strings
texto = "  Hola Mundo Python  "
print("Original:", texto)
print("Slicing [2:6]:", texto[2:6])  # Extrae 'Hola'
print("Slicing [-6:]:", texto[-6:])  # Extrae 'Python'

# Métodos de minúsculas y mayúsculas
print("Minúsculas:", texto.lower())
print("Mayúsculas:", texto.upper())
print("Capitalizado:", texto.capitalize())

# Eliminar espacios
print("Sin espacios al inicio/fin:", texto.strip())
print("Sin espacios a la izquierda:", texto.lstrip())
print("Sin espacios a la derecha:", texto.rstrip())

# Reemplazar texto
nuevo_texto = texto.replace("Mundo", "Universo")
print("Reemplazado:", nuevo_texto)
```

```python
# Ejemplo de interpolación usando f-strings
mensaje = f"El texto original es: '{texto}' y el texto modificado es: '{nuevo_texto}'"
print(mensaje)
```

```python
# Ejemplo de uso de caracteres especiales
ejemplo_caracteres = f"Texto original:\n{texto}\nTexto modificado:\t{nuevo_texto}"
print(ejemplo_caracteres)
```

# Entrada y Salida

```python
# Crear una lista para almacenar los registros
registros = []

while True:
    nombre_usuario = input("Ingrese su nombre (o escriba 'fin' para terminar): ")
    if nombre_usuario.lower() == 'fin':
        break
    edad_usuario = input("Ingrese su edad: ")
    ciudad_usuario = input("Ingrese su ciudad: ")
    registros.append({
        "Nombre": nombre_usuario,
        "Edad": edad_usuario,
        "Ciudad": ciudad_usuario
    })

# Mostrar todos los registros ingresados
for registro in registros:
    print("Nombre:", registro["Nombre"], "Edad:", registro["Edad"], "Ciudad:", registro["Ciudad"], sep=" | ", end=" <-- Fin de datos\n")
```
