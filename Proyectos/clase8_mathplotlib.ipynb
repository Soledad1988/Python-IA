#### üì¶ 4. Boxplot (diagrama de caja)

Caja (entre Q1 y Q3)
- Extremos de la caja:
- Q1 (cuartil inferior): el 25% de los datos est√°n por debajo de este valor.
- Q3 (cuartil superior): el 75% de los datos est√°n por debajo de este valor.
- Altura de la caja:
- Es la rango intercuart√≠lico (IQR) = Q3 ‚àí Q1.
- Mide la dispersi√≥n central, √∫til para detectar variabilidad sin que los extremos influyan.

3. üìè Bigotes (l√≠neas que salen de la caja)
- Extienden hasta los valores m√°s extremos que no se consideran outliers.
- Por defecto en Matplotlib:
- L√≠mite inferior: Q1 ‚àí 1.5 √ó IQR
- L√≠mite superior: Q3 + 1.5 √ó IQR
- Los datos dentro de estos l√≠mites se consideran "normales".
- Valores fuera de los bigotes se marcan como puntos individuales: outliers.
  
üß† Ejemplo pr√°ctico
Supongamos que est√°s analizando la duraci√≥n de turnos m√©dicos:
- La l√≠nea central te dice cu√°l es la duraci√≥n t√≠pica del turno.
- Los bigotes te muestran el rango de duraciones que son razonables.
- Si hay puntos fuera de los bigotes, podr√≠an indicar errores de carga, turnos at√≠picos o condiciones especiales (como guardias extendidas).



```python
datos = [7, 15, 13, 18, 21, 22, 23, 25, 30, 35, 49]

plt.boxplot(datos)
plt.title('Boxplot')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```

#### üßµ 7. Gr√°fico de √°reas (stackplot)
Un stackplot muestra varias series de datos como √°reas apiladas una sobre otra. Cada √°rea representa una categor√≠a o grupo, y su altura acumulada indica el total en cada punto del eje X.

Imaginemos que est√°s analizando la cantidad de horas trabajadas por tres equipos durante una semana:
üîç Lectura visual
- El √°rea inferior (Equipo A) se apila primero.
- Encima se dibuja el √°rea del Equipo B, sumando sus valores a los del A.
- Finalmente, el √°rea del Equipo C se apila encima de las dos anteriores.
- La altura total en cada d√≠a representa la suma acumulada de horas trabajadas por los tres equipos.

üß† Aplicaciones pr√°cticas
- üîã Energ√≠a: consumo por tipo de fuente (solar, e√≥lica, t√©rmica).
- üè• Turnos m√©dicos: horas por especialidad o sector.
- üì¶ Producci√≥n: unidades por l√≠nea de fabricaci√≥n.
- üìà Finanzas: ingresos por categor√≠a de producto.
- üìö Educaci√≥n: participaci√≥n por grupo en actividades


```python
dias = [1, 2, 3, 4, 5]
equipo_A = [2, 3, 4, 3, 2]
equipo_B = [1, 2, 1, 2, 1]
equipo_C = [3, 2, 2, 3, 4]

plt.stackplot(dias, equipo_A, equipo_B, equipo_C,
              labels=['Equipo A', 'Equipo B', 'Equipo C'],
              colors=['#FF9999', '#99CCFF', '#99FF99'])
plt.legend(loc='upper left')
plt.title('Horas trabajadas por equipo')
plt.xlabel('D√≠a')
plt.ylabel('Horas acumuladas')
plt.show()
```

#### üß≠ 8. Gr√°fico polar
Un gr√°fico polar representa datos en coordenadas polares, donde cada punto se define por:
- Œ∏ (theta): el √°ngulo en radianes (eje circular).
- r: la distancia desde el centro (radio).
En lugar de usar un eje X lineal, el gr√°fico se dibuja alrededor de un c√≠rculo, lo que lo hace ideal para visualizar fen√≥menos peri√≥dicos, direccionales o rotacionales.

üìä ¬øC√≥mo se interpreta?
- El eje angular (Œ∏) se representa como una circunferencia completa (0 a 2œÄ radianes o 0¬∞ a 360¬∞).
- El eje radial (r) va desde el centro hacia afuera.
- Cada punto se ubica seg√∫n su √°ngulo y distancia al centro.
- Las curvas que se forman pueden mostrar patrones c√≠clicos, simetr√≠as o direcciones dominantes.

üß† Aplicaciones pr√°cticas
- Meteorolog√≠a: Direcci√≥n y velocidad del viento üå¨Ô∏è
- Ingenier√≠a: An√°lisis de vibraciones o fuerzas radiales ‚öôÔ∏è
- Navegaci√≥n: Trayectorias de radares o sonares üß≠
- Medicina: Distribuci√≥n angular de se√±ales EEG/ECG üß†
- Marketing: Preferencias por segmentos horarios o zonas üõçÔ∏è



```python
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(theta))

plt.polar(theta, r)
plt.title('Gr√°fico polar')
plt.show()
```

```python

# Datos simulados: direcci√≥n en radianes y velocidad
direcciones_radianes = np.array([0, np.pi/6, np.pi/3, np.pi/2, 2*np.pi/3, 5*np.pi/6,
                                 np.pi, 7*np.pi/6, 4*np.pi/3, 3*np.pi/2, 5*np.pi/3, 11*np.pi/6])
velocidades = np.array([5, 7, 3, 6, 8, 4, 2, 5, 6, 7, 3, 4])

# Crear gr√°fico polar
plt.figure(figsize=(8, 6))
plt.polar(direcciones_radianes, velocidades, marker='o', linestyle='-', color='darkorange')
plt.title('Direcci√≥n e intensidad del viento (polar)', fontsize=14)
plt.show()
```

#### üß© 9. Subplots (m√∫ltiples gr√°ficos)
