### Ejemplos con *for*

```python
# Recorrer una lista de temperaturas

temperaturas = [18.5, 21.0, 19.8, 22.3, 17.9]

for t in temperaturas:
    print(f"Temperatura registrada: {t}Â°C")
```

```python
# Iterar sobre un diccionario de sensores

sensores = {
    "Sensor A": "OK",
    "Sensor B": "ERROR",
    "Sensor C": "OK"
}

for nombre, estado in sensores.items():
    print(f"{nombre}: Estado = {estado}")
```

```python
# Filtrar elementos en una lista

valores = [12, -3, 0, 7, -8, 15]
positivos = []

for v in valores:
    if v < 0:
        positivos.append(v)

print("Valores positivos:", positivos)
```

```python
# Iterar sobre un conjunto

materiales = {"acero", "plÃ¡stico", "vidrio"}

for m in materiales:
    print(f"Material inspeccionado: {m}")
```

## Ejemplos con *while*

```python
# Contador descendente

contador = 5

while contador > 0:
    print(f"Cuenta regresiva: {contador}")
    contador -= 1

print("Â¡Despegue!")
```

```python
# ValidaciÃ³n de entrada numÃ©rica

entrada = input("Ingresa un nÃºmero positivo: ")

while not entrada.isdigit() or int(entrada) <= 0:
    entrada = input("Entrada invÃ¡lida. Intenta de nuevo: ")

print(f"NÃºmero vÃ¡lido: {entrada}")
```

```python
# SimulaciÃ³n de carga de baterÃ­a

bateria = 10

while bateria < 100:
    print(f"Cargando... baterÃ­a al {bateria}%")
    bateria += 5

print("BaterÃ­a completamente cargada ðŸ”‹")
```

```python
# ExtracciÃ³n de elementos de una lista

tareas = ["medir presiÃ³n", "verificar voltaje", "calibrar sensor"]

while tareas:
    tarea_actual = tareas.pop(0)            # Extrae la primera tarea, la elimina de la lista
    print(f"Ejecutando tarea: {tarea_actual}")
```
