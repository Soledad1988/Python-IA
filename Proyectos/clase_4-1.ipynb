## Condicionales

### Condicionales Clasicos

```python
# Clasificacion de Temperatura

temp = float(input("Ingresa la temperatura en ¬∞C: "))

if temp < 0:
    print("Est√° congelado ‚ùÑÔ∏è")
elif temp < 20:
    print("Hace fr√≠o üß•")
elif temp < 30:
    print("Clima templado üå§Ô∏è")
else:
    print("Hace calor üî•")
```

### Condicionales Anidados

```python
# Validacion de usuario y rol

usuario = input("Usuario: ")
rol = input("Rol (admin/usuario): ")

if usuario == "Luis":
    if rol == "admin":
        print("Acceso total concedido.")
    else:
        print("Acceso limitado.")
else:
    print("Usuario no reconocido.")
```

## Con estructura de datos

```python
# Verificacion de permisos

permisos = {
    "Luis": ["leer", "escribir", "modificar"],
    "Ana": ["leer"],
    "Carlos": ["leer", "escribir"]
}

usuario = input("Usuario: ")
accion = input("Acci√≥n: ")

if usuario not in permisos:
    print(usuario + " Usuario no existe ‚ùå")
elif accion in permisos[usuario]:
    print(usuario, permisos[usuario])
    print("Acci√≥n permitida ‚úÖ")
else:
    print(usuario, "[" + accion + "]")
    print("Acci√≥n denegada ‚ùå")
```

## Condicionales con funciones

```python
# Evaluacion de notas

def evaluar_nota(nota):
    if nota <  0 or nota > 10:
        return "Nota inv√°lida"
    elif nota >= 9:
        return "Excelente"
    elif nota >= 7:
        return "Bueno"
    elif nota >= 4:
        return "Regular"
    else:
        return "Insuficiente"

# Caso de uso
nota = float(input("Ingresa la nota: "))
print(nota, evaluar_nota(nota))
```

## Utilizando *break*, *continue* y *pass*

```python
# Escaneo de sensores en una linea de producci√≥n

sensores = ["OK", "NOCONFIG", "OK", "NO_CONFIG", "OK", "OK", "NO_CONFIG", "OK", "ERROR", "OK"]

for i, estado in enumerate(sensores):
    print(f"Sensor {i+1}: ", end="")

    if estado == "ERROR":
        print("‚ö†Ô∏è Error detectado. Deteniendo escaneo.")
        break  # Detiene el bucle ante un error cr√≠tico

    elif estado == "NO_CONFIG":
        print("üõ†Ô∏è Sensor sin configurar. Saltando...")
        continue  # Salta esta iteraci√≥n y sigue con el siguiente sensor

    elif estado == "OK":
        pass  # No hace nada especial, pero se puede usar como marcador de lugar
        print("‚úÖ Todo en orden.")

    else:
        print("üõ†Ô∏è Sensor en estado desconocido, revisar. Saltando...")
        continue  # Salta esta iteraci√≥n y sigue con el siguiente sensor        

print("Escaneo finalizado.")
```
