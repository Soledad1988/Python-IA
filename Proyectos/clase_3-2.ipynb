# Colecciones

## Listas

```python
# Definición de una lista / Defining a list
numeros = [10, 20, 30, 40, 50]

# Mostrar la lista completa / Display the full list
print("Lista completa:", numeros)

# Acceder a elementos individuales / Access individual elements
print("Primer elemento:", numeros[0])        # 10
print("Último elemento:", numeros[-1])       # 50

# Agregar un nuevo elemento / Add a new element
numeros.append(60)
print("Lista después de agregar 60:", numeros)

# Eliminar un elemento / Remove an element
numeros.remove(30)
print("Lista después de eliminar 30:", numeros)

# Recorrer la lista con un bucle / Loop through the list
print("Recorriendo la lista:")
for numero in numeros:
    print(numero)

# Obtener la longitud de la lista / Get the length of the list
print("Cantidad de elementos:", len(numeros))
```

## Tuplas

```python
# Definición de una tupla / Defining a tuple
coordenadas = (12.5, -8.3)

# Mostrar la tupla completa / Display the full tuple
print("Tupla completa:", coordenadas)

# Acceder a elementos individuales / Access individual elements
print("Primera coordenada:", coordenadas[0])   # 12.5
print("Segunda coordenada:", coordenadas[1])   # -8.3

# Recorrer la tupla con un bucle / Loop through the tuple
print("Recorriendo la tupla:")
for valor in coordenadas:
    print(valor)

# Obtener la longitud de la tupla / Get the length of the tuple
print("Cantidad de elementos:", len(coordenadas))

# Intentar modificar un elemento (esto generará error) / Try to modify an element (will raise error)
# coordenadas[0] = 0.0  # Descomentar esta línea para ver el error / Uncomment to see the error
```

## Sets (Cpmkimtps)

```python
# Definición de un conjunto / Defining a set
colores = {"rojo", "verde", "azul"}

# Mostrar el conjunto completo / Display the full set
print("Conjunto completo:", colores)

# Agregar un nuevo elemento / Add a new element
colores.add("amarillo")
print("Después de agregar 'amarillo':", colores)

# Eliminar un elemento / Remove an element
colores.discard("verde")
print("Después de eliminar 'verde':", colores)

# Recorrer el conjunto con un bucle / Loop through the set
print("Recorriendo el conjunto:")
for color in colores:
    print(color)

# Verificar si un elemento está presente / Check if an element is present
print("¿'azul' está en el conjunto?", "azul" in colores)

# Operaciones de conjuntos / Set operations
otros_colores = {"azul", "negro", "blanco"}
union = colores.union(otros_colores)
interseccion = colores.intersection(otros_colores)

print("Unión de conjuntos:", union)
print("Intersección de conjuntos:", interseccion)
```

## Dict (Diccionarios)

```python
# Definición de un diccionario con tres registros / Defining a dictionary with three records
estudiantes = {
    "001": {"nombre": "Ana", "edad": 21, "carrera": "Ingeniería"},
    "002": {"nombre": "Luis", "edad": 22, "carrera": "Matemática"},
    "003": {"nombre": "Sofía", "edad": 20, "carrera": "Física"}
}

# Mostrar el diccionario completo / Display the full dictionary
print("Diccionario de estudiantes:")
for legajo, datos in estudiantes.items():
    print(f"Legajo {legajo}: {datos}")

# Acceder a un registro específico / Access a specific record
print("\nDatos del estudiante con legajo '002':")
print(estudiantes["002"])

# Agregar un nuevo registro / Add a new record
estudiantes["004"] = {"nombre": "Carlos", "edad": 23, "carrera": "Química"}
print("\nDespués de agregar el legajo '004':")
print(estudiantes["004"])

# Modificar un valor dentro de un registro / Modify a value inside a record
estudiantes["001"]["edad"] = 22
print("\nEdad actualizada de Ana:")
print(estudiantes["001"])

# Eliminar un registro / Remove a record
del estudiantes["003"]
print("\nDespués de eliminar el legajo '003':")
for legajo, datos in estudiantes.items():
    print(f"Legajo {legajo}: {datos}")
```

## Caracteristicas de las Colecciones

| Tipo de colección | Sintaxis         | Mutabilidad | Ordenado | Permite duplicados | Acceso por índice | Uso típico                                      |
|-------------------|------------------|-------------|----------|---------------------|-------------------|------------------------------------------------|
| `list` (lista)    | `[1, 2, 3]`       | ✅ Sí       | ✅ Sí    | ✅ Sí               | ✅ Sí             | Secuencia de elementos modificables            |
| `tuple` (tupla)   | `(1, 2, 3)`       | ❌ No       | ✅ Sí    | ✅ Sí               | ✅ Sí             | Datos constantes o protegidos contra cambios   |
| `set` (conjunto)  | `{1, 2, 3}`       | ✅ Sí       | ❌ No    | ❌ No               | ❌ No             | Agrupar elementos únicos, operaciones de conjunto |
| `dict` (diccionario) | `{"a": 1, "b": 2}` | ✅ Sí   | ✅ Sí    | ❌ Claves únicas     | ✅ Por clave      | Asociación clave-valor, estructuras jerárquicas |


- **Sintaxis**: Forma en que se define cada tipo de colección en Python. Por ejemplo, listas usan corchetes `[ ]`, tuplas paréntesis `( )`, sets llaves `{ }`, y diccionarios también llaves `{ }` pero con pares `clave: valor`.

- **Mutabilidad**: Indica si los elementos de la colección pueden modificarse después de su creación. Las listas, sets y diccionarios son mutables; las tuplas no.

- **Ordenado**: Señala si los elementos mantienen el orden en que fueron definidos. Las listas, tuplas y diccionarios (desde Python 3.7+) sí lo hacen; los sets no.

- **Permite duplicados**: Define si se pueden repetir elementos dentro de la colección. Las listas y tuplas permiten duplicados; los sets y las claves de los diccionarios no.

- **Acceso por índice**: Indica si se puede acceder a los elementos usando índices numéricos. Solo listas y tuplas permiten acceso por índice; los sets no, y los diccionarios usan claves.

- **Uso típico**: Describe el propósito común de cada colección. Por ejemplo:
  - Listas: secuencias modificables.
  - Tuplas: datos constantes.
  - Sets: agrupación sin duplicados.
  - Diccionarios: asociación clave-valor.
