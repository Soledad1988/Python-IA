### Seaborn (sns)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

#### 1. 游늳 Gr치fico de l칤neas
> Uso: evoluci칩n de variables en el tiempo (producci칩n, temperatura, ventas).

```python
# Tendencias temporales - Gr치fico de l칤neas - lineplot()
df = pd.DataFrame({'d칤a': [1, 2, 3, 4, 5], 'temperatura': [22, 24, 23, 25, 26]})
sns.lineplot(data=df, x='d칤a', y='temperatura')
plt.title('Temperatura diaria')
plt.show()
```

#### 2. 游늴 Gr치fico de dispersi칩n
> Uso: relaci칩n entre dos variables cuantitativas (altura vs peso, ventas vs publicidad); correlaciones, agrupaciones por categor칤a.


```python
# Leer dataset 'tips.csv'
df = pd.read_csv('datasets/tips.csv')
df.head()
```

```python
# Relaci칩n entre variables cuantitativas - Gr치fico de dispersi칩n - scatterplot()
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex')
plt.title('Propina vs Total de la cuenta')
plt.show()
```

### 3. 游늵 Gr치fico de barras
> Uso: comparaci칩n de medias por grupo (turnos, sectores, productos).

```python
# Comparaci칩n entre categor칤as - Gr치fico de barras - barplot()
sns.barplot(data=df, x='day', y='total_bill', estimator='mean')
plt.title('Promedio de cuenta por d칤a')
plt.show()
```

### 4. 游빑 Histograma / KDE
> Uso: an치lisis de frecuencia, detecci칩n de sesgos o multimodalidad.

##### 4.1 Histograma de Barras - hisplot()
- Usa histplot() si quer칠s mostrar frecuencias discretas por rangos (ideal para explicar conteo, distribuci칩n, sesgo).
- Frecuencia absoluta o relativa por intervalos
- Cuenta cu치ntos datos caen en cada bin
- Puede usarse junto con kde=True


```python
# Distribuci칩n de datos - Histograma / KDE - histplot()
sns.histplot(data=df, x='total_bill', bins=20, kde=True)
plt.title('Distribuci칩n de montos de cuenta')
plt.show()
```

#### 4.2 Curva de densidad suavizada - kdeplot()
- Usa kdeplot() si quer칠s mostrar la forma general de la distribuci칩n (ideal para detectar multimodalidad, simetr칤a, colas).
- Estimaci칩n continua de la densidad de probabilidad
- Aplica kernel density estimation (KDE)

```python
# Distribuci칩n de datos - KDE - kdeplot()
sns.kdeplot(data=df, x='total_bill', fill=True, alpha=0.5, bw_adjust=0.5, color='blue', label='KDE')
plt.title('Distribuci칩n de montos de cuenta')
plt.show()
```

### 5. 游닍 Boxplot / Violinplot
> Uso: comparaci칩n de dispersi칩n, detecci칩n de outliers.

##### 5.1 Boxplot

```python
# Resumen estad칤stico - Gr치fico de caja / boxplot()
sns.boxplot(data=df, x='day', y='total_bill')
plt.title('Distribuci칩n de cuentas por d칤a')
plt.show()
```

##### 5.2 Violinplot

```python
# Resumen estad칤stico - Gr치fico de violin / violinplot()
sns.violinplot(data=df, x='day', y='total_bill')
plt.title('Distribuci칩n de cuentas por d칤a')
plt.show()
```

#### 6. 游빌 Gr치fico de puntos categ칩ricos
> Uso: visualizar densidad y dispersi칩n dentro de cada grupo.

##### 6.1 Stripplot - gr치fico de franja
Usa stripplot() si quer칠s una visualizaci칩n r치pida y simple, especialmente con muchos datos.  
Por defecto, los puntos pueden solaparse, pero se puede usar jitter=True para dispersarlos aleatoriamente y mejorar la visibilidad.

```python
# Dispersi칩n categ칩rica - stripplot()
sns.stripplot(data=df, x='day', y='tip')
plt.title('Propinas por d칤a')
plt.show()
```

##### 6.2 Swarmpplot - gr치fico de enjambre
Usa swarmplot() para evitar la superposici칩n de puntos, ideal para conjuntos de datos m치s peque침os.  
Te sirve si quer칠s mostrar claramente la distribuci칩n individual sin que los puntos se oculten entre s칤

```python
# Dispersi칩n categ칩rica - Gr치fico de enjambre / swarmplot()
sns.swarmplot(data=df, x='day', y='tip')
plt.title('Propinas por d칤a')
plt.show()
```

#### 7. 游 Mapa de calor (heatmap)
> Uso: an치lisis multivariado, selecci칩n de variables.

```python
# Matriz de correlaci칩n - Mapa de calor / heatmap()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlaciones entre variables num칠ricas')
plt.show()
```

#### 8. 游댕 Gr치fico de regresi칩n
> Uso: modelado lineal, predicci칩n, validaci칩n de hip칩tesis.

##### 8.1 regplot()
- Usa regplot() si quer칠s una visualizaci칩n r치pida y directa entre dos variables.

```python
# Ajuste de regresi칩n - Gr치fico de regresi칩n / regplot()
sns.regplot(data=df, x='total_bill', y='tip')
plt.title('Regresi칩n lineal: propina vs cuenta')
plt.show()
```

##### 8.2 lmplot()
- Usa lmplot() si quer칠s comparar regresiones por grupo, o crear m칰ltiples subgr치ficos con col o row
游 쯈u칠 aporta lmplot()?
- Segmenta autom치ticamente por la variable hue (en este caso, 'sex').
- Traza una l칤nea de regresi칩n para cada grupo.
- Muestra la dispersi칩n de puntos y el ajuste lineal.


```python
# Gr치fico de regresi칩n segmentado por sexo
sns.lmplot(data=df, x='total_bill', y='tip', hue='sex', height=6, aspect=1.2)

# Personalizaci칩n
plt.title('Relaci칩n entre cuenta y propina por sexo')
plt.xlabel('Total de la cuenta ($)')
plt.ylabel('Propina ($)')
plt.show()
```

```python
# Gr치fico de regresi칩n con facetas por d칤a (columnas) y turno (filas)
sns.lmplot(data=df,
           x='total_bill',
           y='tip',
           col='day',
           row='time',
           hue='sex',         # Opcional: segmentar por sexo
           height=4,
           aspect=1.2)

plt.suptitle('Relaci칩n entre cuenta y propina por d칤a y turno', y=1.02)
plt.show()
```

#### 9. 游빏 Gr치fico de pares (pairplot)
> Uso: detecci칩n de patrones, relaciones cruzadas.

```python
# Exploraci칩n multivariada - Gr치fico de pares / pairplot()
sns.pairplot(df[['total_bill', 'tip', 'price_per_person', 'size']], diag_kind='kde')
plt.suptitle('Relaciones entre variables', y=1.02)
plt.show()
```

#### 10. 游빍 Gr치fico de cajas m칰ltiples
> Uso: comparaci칩n entre subgrupos, an치lisis segmentado.

```python
# Gr치ficos categ칩ricos m칰ltiples - Gr치fico de cajas m칰ltiples / catplot()
sns.catplot(data=df, x='day', y='total_bill', kind='box', col='sex')
plt.suptitle('Distribuci칩n de cuentas por sexo y d칤a', y=1.05)
plt.show()
```

#### Ejemplo: viol칤n + swarmplot para ver outliers
Pod칠s usar el par치metro inner='box' para activar el boxplot interno, pero si quer칠s ver los outliers como puntos individuales, lo mejor es superponer un swarmplot() o stripplot() encima del violinplot():


```python
# Gr치fico de viol칤n
sns.violinplot(data=df, x='day', y='total_bill', inner='box', color='lightblue')

# Superponer puntos individuales (incluye outliers)
sns.swarmplot(data=df, x='day', y='total_bill', color='black', size=3)

plt.title('Distribuci칩n de cuentas por d칤a con outliers')
plt.show()
```
