#### Aplanar listas anidadas (con recursividad)

```python
def flatten_list(nested):
    result = []
    for item in nested:
        if isinstance(item, list):
            result.extend(flatten_list(item))  # llamada recursiva agregando elementos
        else:
            result.append(item)
    return result
```

```python
# Caso de uso
anidada = [1, [2, [3, 4], 5], 6, [7, 8], 9] 

desanidada  = flatten_list(anidada)
desanidada
```

#### Agrupar elementos por una clave

```python
# Agrupamiento por la primera letra

def group_by(items, key_func):
    grouped = {}
    for item in items:
        key = key_func(item)
        grouped.setdefault(key, []).append(item)
    return grouped

# Caso de uso
group_by(["apple", "banana", "avocado", "blueberry"], key_func=lambda x: x[0])

# {'a': ['apple', 'avocado'], 'b': ['banana', 'blueberry']}
```

```python
# Agrupamiento por cantidad de letras

def group_by(items, key_func):
    grouped = {}
    for item in items:
        key = key_func(item)
        grouped.setdefault(key, []).append(item)
    return grouped

# Lista de palabras
palabras = ["sol", "luz", "cielo", "estrella", "mar", "nube"]

# Agrupar por cantidad de letras
agrupadas = group_by(palabras, key_func=lambda x: len(x))

print(agrupadas)
```

#### Fibonacci

```python
# Fibonacci con versión recursiva simple

def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

# Imprimir la secuencia hasta el término n
n = 21
for i in range(n):
    print(fib(i), end=' ')


# Caso lento: fib(40)
fib(40)
```

```python
# Fibonacci con memoización

def memo_fib(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        memo[n] = n
    else:
        memo[n] = memo_fib(n-1, memo) + memo_fib(n-2, memo)
    return memo[n]

# Imprimir la secuencia hasta el término n
n = 21
for i in range(n):
    print(memo_fib(i), end=' ')


# Caso lento: fib(40)
memo_fib(40)
```
